### Analysis

- Gender of mice was split evenly between male (51%) and female (49%).
- Capomulin's Summary Statistics were lowever for all other drugs except for Ramicane who had similar: Mean Tumor Volume, Median Tumor Volume, Tumor Volume Variance, Tumor Volume Std. Dev., Tumro Volume Std. Err. 
- Capomulin and Ramicane had highest number of Observed Mouse Time Points.
- Capomulin had no potential outliers.

 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_results_df = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')

# Display the data table for preview
combined_results_df.head()
```

```python
# Checking the number of mice.
mouse_count = combined_results_df['Mouse ID'].nunique()
mouse_count
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = combined_results_df[combined_results_df.duplicated(subset=['Mouse ID', 'Timepoint'])]
print(duplicate_mice['Mouse ID'])
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
combined_results_df.loc[(combined_results_df['Mouse ID'] == 'g989')]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_results_df = combined_results_df.drop(combined_results_df
                                            [combined_results_df['Mouse ID'] == 'g989'].index)
clean_results_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
new_mouse_count = clean_results_df['Mouse ID'].nunique()
new_mouse_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
Mean_Tumor_Volume = clean_results_df[['Drug Regimen', 'Tumor Volume (mm3)']].groupby('Drug Regimen').mean()
Median_Tumor_Volume = clean_results_df[['Drug Regimen', 'Tumor Volume (mm3)']].groupby('Drug Regimen').median()
Variance_Tumor_Volume = clean_results_df[['Drug Regimen', 'Tumor Volume (mm3)']].groupby('Drug Regimen').var()
STD_Tumor_Volume = clean_results_df[['Drug Regimen', 'Tumor Volume (mm3)']].groupby('Drug Regimen').std()
SEM_Tumor_Volume = clean_results_df[['Drug Regimen', 'Tumor Volume (mm3)']].groupby('Drug Regimen').sem()

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
summary_stats_df = pd.DataFrame({
    "Mean Tumor Volume": Mean_Tumor_Volume['Tumor Volume (mm3)'],
    "Median Tumor Volume": Median_Tumor_Volume['Tumor Volume (mm3)'],
    "Tumor Volume Variance": Variance_Tumor_Volume['Tumor Volume (mm3)'],
    "Tumor Volume Std. Dev.": STD_Tumor_Volume['Tumor Volume (mm3)'],
    "Tumor Volume Std. Err.": SEM_Tumor_Volume['Tumor Volume (mm3)']
})
summary_stats_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
agg_summary_stats_df = clean_results_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].agg(
    mean='mean',
    median='median',
    var='var',
    std='std',
    sem='sem')
agg_summary_stats_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
final_four = clean_results_df[clean_results_df['Drug Regimen'].isin(['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])]

# Get the last (greatest) timepoint for each mouse
final_for_last = final_four.groupby('Mouse ID')['Timepoint'].max()

# Merge this group Series with the original DataFrame to get the tumor volume at the last timepoint
final_four_merge = pd.merge(final_for_last, clean_results_df, on=['Mouse ID', 'Timepoint'], how='left')
```

```python
# Put treatments into a list for a for loop (and later for plot labels)
four_treatments = ('Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin')

# Create an empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and determine if there are any potential outliers for each treatment
for treatment in four_treatments:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    subset_data = final_four_merge[final_four_merge['Drug Regimen'] == treatment]['Tumor Volume (mm3)']
    
    # Calculate quartiles and IQR
    quartiles = subset_data.quantile([0.25, 0.5, 0.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    
    # Determine lower and upper bounds for potential outliers
    lower_bound = lowerq - (1.5 * iqr)
    upper_bound = upperq + (1.5 * iqr)
    
    # Identify potential outliers
    outliers = subset_data[(subset_data < lower_bound) | (subset_data > upper_bound)]
    
    # Append the treatment name, quartiles, and potential outliers to tumor_vol_data
    tumor_vol_data.append({
        'Treatment': treatment,
        'Lower Quartile': lowerq,
        'Upper Quartile': upperq,
        'IQR': iqr,
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound,
        'Potential Outliers': outliers.tolist()
    })

# Print the tumor_vol_data list with quartiles and potential outliers
for data in tumor_vol_data:
    print(data)
    

 
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
# Create a list of lists, each containing the tumor volume data for a specific treatment group
data_to_plot = [final_four_merge[final_four_merge['Drug Regimen'] == treatment]['Tumor Volume (mm3)'] for treatment in four_treatments]

plt.boxplot(data_to_plot, labels=four_treatments)
plt.xlabel('Treatment Group')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Distribution of Tumor Volume by Treatment Group')
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
Capomulin = clean_results_df[clean_results_df['Drug Regimen'] == 'Capomulin']
mouse = 'l509'
Capomulin_mouse = Capomulin[Capomulin['Mouse ID'] == mouse]

plt.figure(figsize=(10, 6))  
plt.plot(Capomulin_mouse['Timepoint'], Capomulin_mouse['Tumor Volume (mm3)'], marker='o', linestyle='-', color='b')
plt.xlabel('Timepoint')
plt.ylabel('Tumor Volume (mm3)')
plt.title(f'Tumor Volume vs. Time Point for Mouse ID {mouse} (Capomulin Treatment)')
plt.show()

```

```python
# Generate a scatter plot of average mouse weight vs. the average observed tumor volume for the entire Capomulin regimen

Capomulin = clean_results_df[clean_results_df['Drug Regimen'] == 'Capomulin']
Capomulin_mean = Capomulin.groupby(['Mouse ID']).agg({'Tumor Volume (mm3)': 'mean','Weight (g)': 'mean'})

fig1, ax1 = plt.subplots(figsize=(10, 6)) 
ax1.scatter(Capomulin_mean['Weight (g)'], Capomulin_mean['Tumor Volume (mm3)'], marker='o', color='b')

ax1.set_xlabel('Weight (g)')
ax1.set_ylabel('Average Tumor Volume (mm3)')
ax1.set_title('Mouse Weight vs. Average Tumor Volume for Capomulin Regimen')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
from scipy.stats import linregress

fig1, ax1 = plt.subplots(figsize=(10, 6)) 
ax1.scatter(Capomulin_mean['Weight (g)'], Capomulin_mean['Tumor Volume (mm3)'], marker='o', color='b')

x_values = Capomulin_mean['Weight (g)']
y_values = Capomulin_mean['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope, 2)) + "x + " + str(round(intercept, 2))

plt.plot(x_values, regress_values, "r-")
plt.annotate(line_eq, (20, 36), fontsize=15, color="red")

correlation_coefficient = round(rvalue, 2)
plt.annotate(f"Correlation coefficient (r): {correlation_coefficient}", (20, 35), fontsize=12, color="red")

ax1.set_xlabel('Weight (g)')
ax1.set_ylabel('Average Tumor Volume (mm3)')
ax1.set_title('Mouse Weight vs. Average Tumor Volume for Capomulin Regimen')

# Show the plot 
plt.show()

```
